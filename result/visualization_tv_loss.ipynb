```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV 파일을 읽어 데이터프레임으로 변환
csv_file_path = './avg_loss_tv_list.csv'
df = pd.read_csv(csv_file_path, header=None, names=['Avg Loss TV'])

# 'Avg Loss TV' 열의 값을 숫자로 변환하고 255로 나누어 정규화
df['Normalized Avg Loss TV'] = pd.to_numeric(df['Avg Loss TV'], errors='coerce') / 255.0

# 결과 출력
print(df['Normalized Avg Loss TV'][1:])
```

```python
# 그래프 생성
plt.plot(df['Normalized Avg Loss TV'][1:]/255 * 100)
plt.title('Generator Average Loss % Over Iterations')
plt.xlabel('Iteration')
plt.ylabel('Average Loss TV')
plt.grid(True)
plt.show()
```
